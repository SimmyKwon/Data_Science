```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values over the domain [-4, 4] with fine resolution for smooth plots
x = np.linspace(-4, 4, 1000)

# Compute ReLU: max(0, x)
y_relu = np.maximum(0, x)

# Compute f(x) = a * x^3 + (1 - a) * x^2 for each a
a_values = [0.4, 0.6, 0.8]
y_functions = {}
colors = ['red', 'green', 'orange']
for i, a in enumerate(a_values):
    y = a * x**3 + (1 - a) * x**2
    y_functions[a] = y

# Create the plot
plt.figure(figsize=(7, 5))  # Optional: Set figure size for better visibility

# Plot ReLU
plt.plot(x, y_relu, label='ReLU', color='blue', linewidth=2)

# Plot each f(x)
for a in a_values:
    plt.plot(x, y_functions[a], label=f'f(x) with a={a}', color=colors[a_values.index(a)], linewidth=2)

# Set axis limits for domain and codomain
plt.xlim(-4, 4)
plt.ylim(-4, 4)

# Add labels, title, legend, and grid
plt.xlabel('x')
plt.ylabel('y')
plt.title('ReLU and f(x) = a x^3 + (1 - a) x^2 for a = 0.4, 0.6, 0.8')
plt.legend()
plt.grid(True)

# Display the plot
plt.show()
```
